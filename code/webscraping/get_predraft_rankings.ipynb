Webscraping some FantasyPros pages is annoying, so instead I downloaded the relevant data

```python
import pandas as pd

pd.set_option('display.max_columns', 100)

def process_fantasypros_predraft_rankings(year):
    data = pd.read_csv(f'../../data/fantasypros_predraft_rankings/unprocessed/FantasyPros_{year}_Draft_ALL_Rankings.csv')[['RK', 'PLAYER NAME', 'POS']]

    data.insert(0, 'YEAR', year)

    # grab position rank from string
    data['POSITION RANK'] = data['POS'].str.extract('(\d+)').astype(int)

    # remove position rank from position
    data['POS'] = data['POS'].str.replace('\d+', '', regex=True)

    data.rename(
        columns={
            'RK': 'PREDRAFT RANK',
            'PLAYER NAME': 'PLAYER',
            'POS': 'POSITION',
        },
        inplace=True,
    )

    return data

process_fantasypros_predraft_rankings(2013)
```

```python
from tqdm.autonotebook import trange

df = pd.concat(
    [
        process_fantasypros_predraft_rankings(year)
        for year in trange(2013, 2024)
    ],
    ignore_index=True,
)

df
```

```python
df.to_parquet(r'..\..\data\fantasypros_predraft_rankings\processed\fantasypros_predraft_rankings.parquet')

'Done'
```
