```python
import pandas as pd

pd.set_option('display.max_columns', 100)

df_mod = pd.read_parquet('../../data/model_data/model_data_single_output.parquet')

df_mod
```

```python
import numpy as np

np.random.seed(22)

groups = (df_mod['Season'].astype(str) + ' Week ' + df_mod['Week'].astype(str)).unique()

np.random.shuffle(groups)

groups
```

```python
fold_0 = [i for i in groups[0::3]]

fold_0
```

```python
fold_1 = [i for i in groups[1::3]]

fold_1
```

```python
fold_2 = [i for i in groups[2::3]]

fold_2
```

```python
df_folds = pd.DataFrame(
    {
        'Fold ID': [0 for _ in range(len(fold_0))] + [1 for _ in range(len(fold_1))] + [2 for _ in range(len(fold_2))],
        'Season Week': fold_0 + fold_1 + fold_2,
    },
)

df_folds
```

```python
df_folds.to_parquet('../../data/model_data/folds.parquet')

print('Done')
```
